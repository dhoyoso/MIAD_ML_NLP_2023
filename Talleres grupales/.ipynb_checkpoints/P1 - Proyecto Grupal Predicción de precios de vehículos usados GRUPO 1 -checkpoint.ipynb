```python
from sklearn.preprocessing import OneHotEncoder

columnas_numeric.remove('Price')

XTotal , YTotal = data.drop(columns="Price"), data["Price"]

encoder = OneHotEncoder(sparse_output=False)
encoder.fit(XTotal[list(columnas_categoric)])

XTotal_encoded_vars = encoder.transform(XTotal[list(columnas_categoric)])
dataTesting_encoded_vars = encoder.transform(dataTesting[list(columnas_categoric)])

scaler = StandardScaler()
scaler.fit(XTotal[list(columnas_numeric)])

XTotal_scaled_vars = scaler.transform(XTotal[list(columnas_numeric)])
dataTesting_scaled_vars = scaler.transform(dataTesting[list(columnas_numeric)])

XTotal_encoded_scaled = pd.concat([pd.DataFrame(XTotal_scaled_vars, columns = ['year','mileage']), pd.DataFrame(XTotal_encoded_vars)], axis=1)
dataTesting_encoded_scaled = pd.concat([pd.DataFrame(dataTesting_scaled_vars, columns = ['year','mileage']), pd.DataFrame(dataTesting_encoded_vars)], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(XTotal_encoded_scaled, YTotal, test_size=0.2, random_state=47)
```

```python
# Crear un nuevo modelo de XGB 
xgb_model = xgb.XGBRegressor(tree_method='gpu_hist')

# Entrenar el modelo con los datos de entrenamiento
xgb_model.fit(X_train, y_train)

# Realizar predicciones con el modelo calibrado
y_pred = xgb_model.predict(X_test)  # X_test son los datos de prueba

# Evaluar el desempeo del modelo
mse_xgb = mean_squared_error(y_test, y_pred)  # y_test son las etiquetas verdaderas de los datos de prueba
rmse_xgb = np.sqrt(mse_xgb)
r2_xgb = r2_score(y_test, y_pred)
mae_xgb = mean_absolute_error(y_test, y_pred)
mape_xgb =np.mean(np.abs((y_test - y_pred) / y_test)) * 100

print("Medidas desempeo XGBoost Calibrado MSE: {:.2f}, R2: {:.2f},RMSE: {:.2f}, MAE: {:.2f}, MAPE: {:.2f}".format(mse_xgb, r2_xgb,rmse_xgb, mae_xgb,mape_xgb))
```

```python
# Codificar las variables categricas con la codificacin de frecuencia
encoder = ce.CountEncoder(cols=['State', 'Make', 'Model'])
X = encoder.fit_transform(XTotal)
X.columns

dataTestingEncoded = encoder.transform(dataTesting)

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, YTotal, test_size=0.2, random_state=42)

# Crear un nuevo modelo de XGB 
xgb_model = xgb.XGBRegressor(tree_method='gpu_hist')

# Entrenar el modelo con los datos de entrenamiento
xgb_model.fit(X_train, y_train)

# Realizar predicciones con el modelo calibrado
y_pred = xgb_model.predict(X_test)  # X_test son los datos de prueba

# Evaluar el desempeo del modelo
mse_xgb = mean_squared_error(y_test, y_pred)  # y_test son las etiquetas verdaderas de los datos de prueba
rmse_xgb = np.sqrt(mse_xgb)
r2_xgb = r2_score(y_test, y_pred)
mae_xgb = mean_absolute_error(y_test, y_pred)
mape_xgb =np.mean(np.abs((y_test - y_pred) / y_test)) * 100

print("Medidas desempeo XGBoost Calibrado MSE: {:.2f}, R2: {:.2f},RMSE: {:.2f}, MAE: {:.2f}, MAPE: {:.2f}".format(mse_xgb, r2_xgb,rmse_xgb, mae_xgb,mape_xgb))
```

```python
# Crear un regresor de XGBoost
xgb_model = xgb.XGBRegressor(tree_method='gpu_hist')

# Definir los parmetros y los rangos de bsqueda para la calibracin
# params = {
#     'learning_rate': [0.01, 0.1, 0.2, 0.3], # Valores a probar para learning_rate
#     'gamma': [0, 0.1, 0.2, 0.3, 0.4], # Valores a probar para gamma
#     'colsample_bytree': [0.6, 0.7, 0.8, 0.9] # Valores a probar para colsample_bytree
# }
params = {
    "subsample":[0.75, 1],
    'colsample_bytree': [0.6, 0.7, 0.8, 0.9],
    "max_depth": range (2, 10, 1),
    "min_child_weight":[1, 5],
    "learning_rate":[0.01, 0.1, 0.2, 0.3],
    'gamma': [0, 0.1, 0.2, 0.3, 0.4],
    'n_estimators': range(60, 220, 40)
    }

# Realizar bsqueda de hiperparmetros utilizando validacin cruzada
grid_search = GridSearchCV(xgb_model, params, cv=5)
grid_search.fit(X_train, y_train)

# Imprimir los mejores hiperparmetros encontrados
print("Mejor learning_rate: ", grid_search.best_params_['learning_rate'])
print("Mejor gamma: ", grid_search.best_params_['gamma'])
print("Mejor colsample_bytree: ", grid_search.best_params_['colsample_bytree'])

# Obtener los resultados de la bsqueda de hiperparmetros
results = grid_search.cv_results_

# Crear un grfico de puntos para los resultados de n_estimators
sns.scatterplot(x='param_learning_rate', y='mean_test_score', data=results)
plt.xlabel('learning_rate')
plt.ylabel('Mean Test Score')
plt.title('Grid Search Results - learning_rate')
plt.show()

# Crear un grfico de puntos para los resultados de max_depth
sns.scatterplot(x='param_gamma', y='mean_test_score', data=results)
plt.xlabel('gamma')
plt.ylabel('Mean Test Score')
plt.title('Grid Search Results - gamma')
plt.show()

# Crear un grfico de puntos para los resultados de max_features
sns.scatterplot(x='param_colsample_bytree', y='mean_test_score', data=results)
plt.xlabel('colsample_bytree')
plt.ylabel('Mean Test Score')
plt.title('Grid Search Results - colsample_bytree')
plt.show()

# Obtener los mejores valores de los hiperparmetros
best_learning_rate = grid_search.best_params_['learning_rate']
best_gamma = grid_search.best_params_['gamma']
best_colsample_bytree = grid_search.best_params_['colsample_bytree']

# Crear un nuevo modelo de XGB con los mejores hiperparmetros
best_xgb_model = xgb.XGBRegressor(learning_rate = best_learning_rate, gamma = best_gamma, colsample_bytree = best_colsample_bytree, tree_method='gpu_hist')

# Entrenar el modelo con los datos de entrenamiento
best_xgb_model.fit(X_train, y_train)

# Realizar predicciones con el modelo calibrado
y_pred = best_xgb_model.predict(X_test)  # X_test son los datos de prueba

# Evaluar el desempeo del modelo
mse_best_xgb = mean_squared_error(y_test, y_pred)  # y_test son las etiquetas verdaderas de los datos de prueba
rmse_best_xgb = np.sqrt(mse_best_xgb)
r2_best_xgb = r2_score(y_test, y_pred)
mae_best_xgb = mean_absolute_error(y_test, y_pred)
mape_best_xgb =np.mean(np.abs((y_test - y_pred) / y_test)) * 100

print("Medidas desempeo XGBoost Calibrado MSE: {:.2f}, R2: {:.2f},RMSE: {:.2f}, MAE: {:.2f}, MAPE: {:.2f}".format(mse_best_xgb, r2_best_xgb,rmse_best_xgb, mae_best_xgb,mape_best_xgb))
```
